```python
import docx

# 打开Word文档
doc = docx.Document("GPU显卡安装文档.docx")
```

```python
for paragraph in doc.paragraphs:
    # 获取段落的格式
    paragraph_format = paragraph.paragraph_format
    alignment = paragraph_format.alignment
    indentation_left = paragraph_format.left_indent
    indentation_right = paragraph_format.right_indent

    # 打印段落格式信息
    print("Alignment:", alignment)
    print("Left Indentation:", indentation_left)
    print("Right Indentation:", indentation_right)
```
